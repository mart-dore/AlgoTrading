```python

client = Client(api, secret)

def get_historical_data(symbol, interval, start_str, end_str=None):
    klines = client.get_historical_klines(symbol, interval, start_str, end_str)
    df = pd.DataFrame(klines, columns=['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 
                                       'Close Time', 'Quote Asset Volume', 'Number of Trades', 
                                       'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'])
    df['Open Time'] = pd.to_datetime(df['Open Time'], unit='ms')
    df['Close Time'] = pd.to_datetime(df['Close Time'], unit='ms')
    df['Open'] = df['Open'].astype(float)
    df['High'] = df['High'].astype(float)
    df['Low'] = df['Low'].astype(float)
    df['Close'] = df['Close'].astype(float)
    df['Volume'] = df['Volume'].astype(float)
    return df

# Rcuprer les donnes des dernires 1000 minutes
symbol = 'BTCUSDT'
interval = '1h'
data = get_historical_data(symbol, interval, "100 days ago UTC")

# Calcul des moyennes mobiles
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()

# Dterminer les signaux d'achat/vente
data['Signal'] = 0
data['Signal'][5:] = np.where(data['SMA_5'][5:] > data['SMA_20'][5:], 1, -1)

# Initialisation des colonnes pour les ordres d'achat et de vente
data['Buy'] = np.nan
data['Sell'] = np.nan

# Boucle pour dtecter les moments d'achat et de vente
position = 0  # 0 = pas de position, 1 = acheteur, -1 = vendeur
for i in range(1, len(data)):
    if data['Signal'][i] == 1 and position == 0:  # Acheter
        data['Buy'][i] = data['Close'][i]
        position = 1
    elif data['Signal'][i] == -1 and position == 1:  # Vendre
        data['Sell'][i] = data['Close'][i]
        position = 0


```

```python
# Simuler l'achat et la vente en suivant les croisements
def backtest_strategy(data):
    data['Signal'] = 0  # 0 signifie pas de signal
    data['Signal'][5:] = np.where(data['SMA_5'][5:] > data['SMA_20'][5:], 1, -1)  # 1 = acheter, -1 = vendre

    # Initialiser le portefeuille
    initial_capital = 1000  # en USDT
    position = 0  # nombre de BTC dtenus
    capital = initial_capital  # capital restant en USDT
    
    # Initialisation du capital :
    data['Capital'] = np.nan

    # Boucle sur les donnes pour excuter les signaux de trading
    for i in range(1, len(data)):
        if data['Signal'][i] == 1 and position == 0:  # Acheter si signal d'achat et pas de position ouverte
            position = capital / data['Close'][i]  # Acheter autant de BTC que possible
            capital = 0
            print(f"Achat de {position:.4f} BTC  {data['Close'][i]:.2f} USDT")
        
        elif data['Signal'][i] == -1 and position > 0:  # Vendre si signal de vente et BTC dtenu
            capital = position * data['Close'][i]  # Vendre tout
            print(f"Vente de {position:.4f} BTC  {data['Close'][i]:.2f} USDT, Capital actuel: {capital:.2f} USDT")
            position = 0
        
        data['Capital'][i] = capital

    # Si une position est encore ouverte  la fin, la vendre
    if position > 0:
        capital = position * data['Close'].iloc[-1]
        print(f"Vente de fin de position, Capital final: {capital:.2f} USDT")

    return capital

# Excuter le backtest
final_capital = backtest_strategy(data)
print(f"Capital initial: 1000 USDT, Capital final: {final_capital:.2f} USDT")
```

```python
# Visualisation avec matplotlib et mplfinance
def plot_trading_strategy(data):
    # Prparation des donnes pour mplfinance
    ohlc_data = data[['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume']].copy()
    ohlc_data.set_index('Open Time', inplace=True)
    
    # Crer le graphique des chandeliers avec les SMA
    add_sma_5 = mpf.make_addplot(data['SMA_5'], color='blue', width=1)
    add_sma_20 = mpf.make_addplot(data['SMA_20'], color='yellow', width=1)
    
    # Ajouter les ordres d'achat et de vente
    add_buy_signal = mpf.make_addplot(data['Buy'], type='scatter', marker='^', color='green', markersize=100)
    add_sell_signal = mpf.make_addplot(data['Sell'], type='scatter', marker='v', color='red', markersize=100)

    # Gnrer le graphique final
    mpf.plot(ohlc_data, type='candle', addplot=[add_sma_5, add_sma_20, add_buy_signal, add_sell_signal], 
             title=f'{symbol} Trading Strategy',
             style='charles', volume=False, figsize=(12, 8), ylabel='Price (USDT)')

# Crer et afficher le graphique
plot_trading_strategy(data)
```

```python
import plotly.graph_objects as go
def plot_trading_strategy_interactive(data):
    fig = go.Figure()

    # Ajouter les bougies (candlestick)
    fig.add_trace(go.Candlestick(x=data['Open Time'],
                                 open=data['Open'],
                                 high=data['High'],
                                 low=data['Low'],
                                 close=data['Close'],
                                 name='Candlestick'))

    # Ajouter les moyennes mobiles
    fig.add_trace(go.Scatter(x=data['Open Time'], y=data['SMA_5'], line=dict(color='blue', width=1), name='SMA 5'))
    fig.add_trace(go.Scatter(x=data['Open Time'], y=data['SMA_20'], line=dict(color='yellow', width=1), name='SMA 20'))

    # Ajouter les points d'achat (flches vertes)
    fig.add_trace(go.Scatter(x=data['Open Time'], y=data['Buy'],
                             mode='markers', marker=dict(color='green', symbol='triangle-up', size=10),
                             name='Buy Signal'))

    # Ajouter les points de vente (flches rouges)
    fig.add_trace(go.Scatter(x=data['Open Time'], y=data['Sell'],
                             mode='markers', marker=dict(color='red', symbol='triangle-down', size=10),
                             name='Sell Signal'))

    # Mettre  jour la mise en page du graphique
    fig.update_layout(title=f'{symbol} Trading Strategy',
                      xaxis_title='Time',
                      yaxis_title='Price (USDT)',
                      xaxis_rangeslider_visible=False,
                      template='plotly_dark',
                      width=1000, height=600)

    # Afficher le graphique
    fig.show()

# Afficher le graphique interactif
plot_trading_strategy_interactive(data)
```
